```python
from pathlib import Path
import pandas as pd

file_path = Path("historical_data_2025/historical_data_2025Q1/historical_data_2025Q1.txt")

# Define the exact attribute mapping in the order they appear in the file
origination_columns = [
    "Classic_FICO",                  # Standardized FICO (300-850)
    "First_Payment_Date",            # YYYYMM
    "First_Time_Homebuyer_Indicator",# Y = Yes, N = No
    "Maturity_Date",                 # YYYYMM
    "Metropolitan_Statistical_Area", # 5-digit MSA / Metro Division Code
    "Mortgage_Insurance_Percentage", # MI % (1% - 55%)
    "Number_of_Units",               # 1, 2, 3, or 4 units
    "Occupancy_Status",              # P = Primary, I = Investment, S = Second
    "Original_Combined_LTV",         # CLTV Ratio
    "Original_Debt_to_Income_Ratio", # DTI Ratio (Capped at 65%)
    "Original_UPB",                  # Rounded to the nearest $1,000
    "Original_Loan_to_Value",        # LTV Ratio
    "Original_Interest_Rate",        # Note rate at origination
    "Channel",                       # R = Retail, B = Broker, C = Correspondent
    "Prepayment_Penalty_Indicator",  # Y = Yes, N = No
    "Amortization_Type",             # FRM or ARM
    "Property_State",                # Two-letter state code
    "Property_Type",                 # SF, CO, PU, MH, CP
    "Postal_Code",                   # First 3 digits of postal code
    "Loan_Identifier",               # Unique PYYQnXXXXXXX ID
    "Loan_Purpose",                  # P = Purchase, C = Cash-Out, N = No Cash-Out
    "Original_Loan_Term",            # Number of months scheduled
    "Number_of_Borrowers",           # Count of obligated borrowers
    "Seller_Name",                   # Lender Name or 'OTHER'
    "Servicer_Name",                 # Servicer Name or 'OTHER'
    "Super_Conforming_Flag",         # Y = Yes, N = No
    "Pre_HARP_Loan_Sequence_Number", # Linked loan sequence ID for HARP
    "Special_Eligibility_Program",   # H = Home Possible, F = HFA, R = Refi Possible
    "HARP_Indicator",                # Y = Relief Refinance Loan, N = Non-Relief
    "Property_Valuation_Method",     # 1 = Waiver, 2 = Appraisal, 4 = ACE+ PDR
    "Interest_Only_Indicator",       # Y = Yes, N = No
    "VantageScore_4_0"               # Standardized VantageScore (300-850)
]

# Read the CSV with header=None and assign the clean column names
df = pd.read_csv(
    file_path, 
    sep="|", 
    header=None, 
    names=origination_columns,
    index_col=False
)

# Print the cleanly structured head
df.head()
```
